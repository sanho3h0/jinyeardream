```python
# json 패키지를 임포트합니다.
import json

#JSON 파일을 읽고 문자열을 딕셔너리로 변환합니다.
def create_dict(filename):
    with open(filename) as file:
        json_string = file.read()
        json_dict = json.loads(json_string)
        
        # 함수를 완성하세요.
        return json_dict

#JSON 파일을 읽고 딕셔너리를 JSON 형태의 문자열로 변환합니다.
def create_json(dictionary, filename):
    with open(filename, 'w') as file:
        # 함수를 완성하세요.
        json_string = json.dumps(dictionary)
        file.write(json_string)
        
# 아래 주석을 해제하고 결과를 확인해보세요.  
src = 'netflix.json'
dst = 'new_netflix.json'

netflix_dict = create_dict(src)
print('원래 데이터: ' + str(netflix_dict))

netflix_dict['Dark Knight'] = 39217
create_json(netflix_dict, dst)
updated_dict = create_dict(dst)
print('수정된 데이터: ' + str(updated_dict))
```

```python
set1 = {1, 3, 5, 7}
set2 = {1, 3, 9, 27}

union = set1 | set2 # 합집합 {1, 3, 5, 7, 9, 27}
print(union)

# 교집합
intersection = set1 & set2 # {1, 3}
print(intersection)

# 차집합
diff = set1 - set2 # {5, 7}
print(diff)

# xor 서로 다르면 참 : 두 집합 중 겹치지 않는 요소만 포함
xor = set1 ^ set2 # {5, 7, 9, 27}
print(xor)
```

```python
# csv 모듈을 임포트합니다. 
# filename에 해당하는 파일을 읽기 모드(r)로 엽니다
# with 문을 사용하면 파일을 자동으로 닫아줍니다. (file.close() 필요 없음)
import csv

def print_book_info(filename):
    with open(filename) as file:
        # ',' 기호로 분리된 CSV 파일을 처리하세요..
        reader = csv.reader(file)
        # csv.reader(file)는 file을 읽는 CSV 리더 객체를 생성합니다.
        # reader는 각 줄을 리스트 형태로 반환하는 **반복 가능한 객체(이터레이터)**입니다
        
        # 처리된 파일의 각 줄을 불러옵니다.
        for row in reader:
            # row : csv파일의 각 행을 리스트 형태로 가져온다.
            # row 리스트에서 인덱스를 사용하여 원하는 값을 추출할 수 있습니다.
            
            # 함수를 완성하세요.
            title = row[0] # 책 제목 csv 첫째 열
            author = row[1] #csv 둘째 열
            pages = row[3] #csv 넷째 열
            print("{} ({}): {}p".format(title, author, pages))
            # format()을 사용하여 "제목 (저자) :페이지수p" 형식으로 출력


# 아래 주석을 해제하고 실행 결과를 확인해보세요.
filename = 'books.csv' 
# 파일 디렉토리의 경로를 설정해야함
# 예시 : filename = 'prac_0401/books.csv'
print_book_info(filename)
```

```python
# CSV, JSON 모듈을 임포트합니다.
import csv
import json
# from elice_utils import EliceUtils

# elice_utils = EliceUtils()

def books_to_json(src_file, dst_file):
    # 아래 함수를 완성하세요.
    books = []
    with open(src_file) as src:
        reader = csv.reader(src)
        
        # 각 줄 별로 대응되는 book 딕셔너리를 만듭니다.
        for row in reader:
            # 책 정보를 저장하는 딕셔너리를 생성합니다.
            book = {
                "title" : row[0],
                "author" : row[1],
                "genre" : row[2],
                "pages" : int(row[3]),
                "publisher" : row[4]
            }
            books.append(book)
    
    with open(dst_file, 'w') as dst:
        # 딕셔너리 -> 문자열 (json)
        book_str = json.dumps(books)
        # JSON 형식으로 dst_file에 저장합니다.
        dst.write(book_str)


# 아래 주석을 해제하고 결과를 확인해보세요.  
src_file = 'books.csv'
dst_file = 'books.json'
books_to_json(src_file, dst_file)
# elice_utils.send_file(dst_file)

# dumps하고 write대신에 한번에
# json.dump(books, dst)로 해도 될까요?
```

```python
'''
주어진 값이 정수가 아니거나 최솟값 minimum보다 작으면 False를 리턴하는 함수를 리턴합니다.

'''
def min_validator(minimum):
    def helper(n):
        # n의 타입이 정수가 아니면 False를 리턴합니다.
        if type(n) is not int:
            return False
        
        # 아래 함수를 완성하세요.
        return n >= minimum
    
    return helper
'''    
주어진 값이 정수가 아니거나 최댓값 maximum보다 크면 False를 리턴하는 함수를 리턴합니다.
'''
def max_validator(maximum):
    def helper(n):
        # n의 타입이 정수가 아니면 False를 리턴합니다.
        if type(n) is not int:
            return False
        
        # 아래 함수를 완성하세요.
        return n <= maximum
    
    return helper


def validate(n, validators):
    # validator 중 하나라도 통과하지 못하면 False를 리턴합니다.
    for validator in validators:
        if not validator(n):
            return False
    
    return True


# 작성한 함수를 테스트합니다. # 아래 주석을 해제하고 결과 값을 확인해보세요.
# # 나이 데이터를 검증하는 validator를 선언합니다. 
age_validators = [min_validator(0), max_validator(120)]
ages = [9, -3, 7, 33, 18, 1999, 287, 0, 13]

# 주어진 나이 데이터들에 대한 검증 결과를 출력합니다.
print("검증 결과")
for age in ages:
    result = "유효함" if validate(age, age_validators) else "유효하지 않음"
    print("{}세 : {}".format(age, result))
```

```python
# CSV 모듈을 임포트합니다.
import csv

def get_titles(books_csv):
    with open(books_csv) as books:
        reader = csv.reader(books, delimiter=',')
        # 함수를 완성하세요.
        get_title = lambda row: row[0]
        titles = map(get_title,reader)
        
        return list(titles)


# 작성한 코드를 테스트합니다. 주석을 해제하고 실행하세요.
books = 'books.csv'
titles = get_titles(books)
for title in titles:
     print(title)
```
